## Parte 2 de PCA

En este notebook, hacemos dos ejemplos simulados rapido que represetan contextos de **baja** y **alta correlacion** entre los predictores y la variabilidad explicada de los primeros componentes principales de PCA.

```python
# Importamos los paquetes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

### 1. Contexto de baja correlación de X 
Primero, simulamos tener 4 variables con correlacion lineal de casi 0.

```python
# Establecer semilla para reproducibilidad
np.random.seed(42)

# Simulación de 4 variables no correlacionadas
n_samples = 200

X1 = np.random.normal(loc=100, scale=15, size=n_samples)
X2 = np.random.normal(loc=50, scale=10, size=n_samples)
X3 = np.random.normal(loc=75, scale=20, size=n_samples)
X4 = np.random.normal(loc=25, scale=5, size=n_samples)

# Crear DataFrame
datos = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3,
    'X4': X4
})
```

Ahora visualizamos dichos datos creados y baja correlación entre los mismos

```python
print("Primeras filas de los datos:")
print(datos.head().round(3))
print(f"\nShape: {datos.shape}")
```

```python
datos.describe().round(2)
```

```python
corr_original = datos.corr()

# Visualizar correlaciones originales
plt.figure(figsize=(8, 6))
sns.heatmap(corr_original, annot=True, cmap='coolwarm_r', center=0, 
            fmt='.2f', square=True, vmin=-1, vmax=1)
plt.title('Matriz de Correlaciones Original\n(Variables No Correlacionadas)', 
          fontsize=12)
plt.tight_layout()
plt.show()
```

Vemos correlaciones muy bajitas de casi 0. Ahora aplicamos la estandarización a las variables y luego PCA.

```python
# Estandarizar los datos (importante para PCA)
scaler = StandardScaler()
datos_estandarizados = scaler.fit_transform(datos)

# Aplicar PCA
pca = PCA()
componentes_principales = pca.fit_transform(datos_estandarizados)

# Crear DataFrame con los componentes principales
pcs_df = pd.DataFrame(
    componentes_principales,
    columns=[f'PC{i+1}' for i in range(4)]
)
```

```python
#Creamos la proporciones de varianza explicada y acumulada
varianza_explicada = pca.explained_variance_ratio_
varianza_acumulada = np.cumsum(varianza_explicada)

for i, (var, var_acum) in enumerate(zip(varianza_explicada, varianza_acumulada)):
    print(f"PC{i+1}: {var*100:.2f}% | Acumulada: {var_acum*100:.2f}%")

# Loadings (correlaciones entre variables originales y PCs)
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(4)],
    index=['X1', 'X2', 'X3', 'X4']
)
```

```python
# Gráfico de varianza explicada (Scree Plot)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(range(1, 5), varianza_explicada*100, alpha=0.7, color='steelblue', edgecolor='black')
plt.xlabel('Componente Principal', fontsize=11)
plt.ylabel('Varianza Explicada (%)', fontsize=11)
plt.title('Varianza por Componente', fontsize=12)
plt.xticks(range(1, 5))

# Varianza acumulada
plt.subplot(1, 2, 2)
plt.plot(range(1, 5), varianza_acumulada*100, marker='o', linewidth=2, markersize=8, color='darkgreen')
plt.xlabel('Componente Principal', fontsize=11)
plt.ylabel('Varianza Acumulada (%)', fontsize=11)
plt.title('Varianza Acumulada', fontsize=12)
plt.xticks(range(1, 5))
plt.ylim([0, 105])
plt.tight_layout()
plt.show()
```

### 2. Contexto de ALTA correlación entre las Xs

```python
# Establecer semilla para reproducibilidad
np.random.seed(42)

# Simulación de 4 variables ALTAMENTE CORRELACIONADAS
n_samples = 200

# Crear una variable latente (factor común)
factor_comun = np.random.normal(loc=0, scale=1, size=n_samples)

# Las 4 variables se derivan del factor común con ruido mínimo
X1 = 100 + 15 * factor_comun + np.random.normal(0, 2, n_samples)
X2 = 50 - 10 * factor_comun + np.random.normal(0, 1.5, n_samples)
X3 = 75 - 20 * factor_comun + np.random.normal(0, 2.5, n_samples)
X4 = 25 + 5 * factor_comun + np.random.normal(0, 1, n_samples)

# Crear DataFrame
datos = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3,
    'X4': X4
})

print("Primeras filas de los datos:")
print(datos.head().round(2))
print(f"\nShape: {datos.shape}")

```

```python
# Matriz de correlación original
corr_original = datos.corr()

# Visualizar correlaciones originales
plt.figure(figsize=(8, 6))
sns.heatmap(corr_original, annot=True, cmap='coolwarm_r', center=0, 
            fmt='.3f', square=True, vmin=-1, vmax=1, cbar_kws={'label': 'Correlación'})
plt.title('Matriz de Correlaciones Original\n(Variables Altamente Correlacionadas)', 
          fontsize=12)
plt.tight_layout()
plt.show()
```

```python

# Estandarizar los datos (importante para PCA)
scaler = StandardScaler()
datos_estandarizados = scaler.fit_transform(datos)

# Aplicar PCA
pca = PCA()
componentes_principales = pca.fit_transform(datos_estandarizados)

# Crear DataFrame con los componentes principales
pcs_df = pd.DataFrame(
    componentes_principales,
    columns=[f'PC{i+1}' for i in range(4)]
)

```

```python
#VARIANZA EXPLICADA POR CADA COMPONENTE
varianza_explicada = pca.explained_variance_ratio_
varianza_acumulada = np.cumsum(varianza_explicada)

for i, (var, var_acum) in enumerate(zip(varianza_explicada, varianza_acumulada)):
    print(f"PC{i+1}: {var*100:.2f}% | Acumulada: {var_acum*100:.2f}%")
```

```python
# Gráfico de varianza explicada (Scree Plot)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
colores = ['darkgreen' if i == 0 else 'steelblue' for i in range(4)]
barras = plt.bar(range(1, 5), varianza_explicada*100, alpha=0.7, color=colores, edgecolor='black', linewidth=1.5)
plt.xlabel('Componente Principal', fontsize=11)
plt.ylabel('Varianza Explicada (%)', fontsize=11)
plt.title('Varianza por Componente\n(Concentración en PC1)', fontsize=12)
plt.xticks(range(1, 5), fontsize=11)

# Varianza acumulada
plt.subplot(1, 2, 2)
plt.plot(range(1, 5), varianza_acumulada*100, marker='o', linewidth=3, markersize=10, 
         color='darkgreen', markerfacecolor='lightgreen', markeredgewidth=2, markeredgecolor='darkgreen')
plt.fill_between(range(1, 5), varianza_acumulada*100, alpha=0.3, color='green')
plt.xlabel('Componente Principal', fontsize=11)
plt.ylabel('Varianza Acumulada (%)', fontsize=11)
plt.title('Varianza Acumulada\n(Dominancia de Primeros Componentes)', fontsize=12)
plt.xticks(range(1, 5), fontsize=11)
plt.ylim([0, 105])
plt.grid(alpha=0.3)
# Añadir línea de referencia al 95%
plt.axhline(y=95, color='red', linestyle='--', linewidth=2, alpha=0.7, label='95% varianza')
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

#### Diferencias clave entre correlación alta vs. baja:

| Aspecto | Correlación Baja | Correlación Alta |
|--------|---|---|
| **Varianza en PC1** | ~25% | ~95% |
| **Scree Plot** | Gradual | Concentrada en PC1 |
| **Biplot** | Vectores dispersos | Vectores en misma dirección |
| **Loadings PC1** | Pequeños y variados | Grandes y similares |
| **Reducción dim.** | Necesito 3-4 PCs | Basta con 1-2 PCs |
| **Correlación original** | Cercana a 0 | Cercana a 1 |

**Interpretación:**
- Cuando las variables están altamente correlacionadas, el PCA las resume en muy pocos componentes
- El PC1 actúa como un "factor común" que captura la variación conjunta
- Esto es útil para reducción de dimensionalidad sin perder información relevante
